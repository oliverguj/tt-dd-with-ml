`````python
# read the OneCore currated data
def read_onecore_json(device_type= 'phone'):
    oc_test_data = pd.read_json('userAgents.json')
    if device_type == 'phone':
        y_oc_test = oc_test_data['expected'].apply(lambda row: row['isMobile']) 
    elif device_type == 'tablet':
        y_oc_test = oc_test_data['expected'].apply(lambda row: row['isTablet']) 
    elif device_type == 'computer':
        y_oc_test = oc_test_data['expected'].apply(lambda row: not row['isMobile'] and not row['isTablet']) 
    else:
        y_oc_test = None

    oc_test_ua = oc_test_data['userAgent']

    return (oc_test_ua, y_oc_test)
`````

`````python
def read_whatsmybrowser_csv(device_type= 'phone'):
    # https://developers.whatismybrowser.com/useragents/database/
    data_wmb = pd.read_csv('whatismybrowser-user-agent-database.csv')
    # print(data_wmb.keys())
    # print(data_wmb['hardware_type'].unique())
    # print(data_wmb['hardware_sub_type'].unique())
    # print(data_wmb['hardware_sub_sub_type'].unique())

    data_wmb = data_wmb[data_wmb['hardware_type'] != 'server']

    user_agents = data_wmb['user_agent']

    if device_type == 'computer':
        y = data_wmb['hardware_type'] == device_type
    else:
        y = data_wmb['hardware_sub_type'] == device_type

    return (user_agents, y)
`````

`````python
def describe_sets(X_train, X_test, y_train, y_test):
    print('Number of rows in the total set: {}'.format(X.shape[0]))
    print('Number of rows in the training set: {}'.format(X_train.shape[0]))
    print('Number of rows in the test set: {}'.format(X_test.shape[0]))
    print('Number of {} devices in training set: {:.0f}'.format(device_type, np.sum(y_train * np.ones(np.shape(y_train)))))
    print('Number of {} devices in test set: {:.0f}'.format(device_type, np.sum(y_test * np.ones(np.shape(y_test)))))

describe_sets(X_train, X_test, y_train, y_test)
`````

`````python
from sklearn.feature_extraction.text import CountVectorizer

def create_stop_word_list(X):
    count_vector = CountVectorizer(binary = True, lowercase=False, max_df=0.9, min_df=5)
    count_vector.fit(X)

    # find numbers and some manually added stopwords
    import re
    stop_words = [] # ['cvcv', 'RuxitSynthetic', 'smf', 'X11'] #'sm']
    stop_words_regex = re.compile('^[0-9]*$')
    for word in count_vector.get_feature_names_out():
        if stop_words_regex.match(word):
            stop_words.append(word)
    
    return stop_words

stop_words = create_stop_word_list(X)
stop_words[0:10]
`````

`````python
# build a dictionary of all words (except the stop_words)

# https://towardsdatascience.com/basics-of-countvectorizer-e26677900f9c

count_vector = CountVectorizer(binary = True, ngram_range=[1,1], lowercase=False, max_df=0.9, min_df=5, stop_words=stop_words, max_features=100)
count_vector.fit(X)
print('{} words in feature set'.format(np.count_nonzero(count_vector.get_feature_names_out())))
print('{} first features: {}'.format(30, count_vector.get_feature_names_out()[0:30]))
print('stop-words list: {}'.format(count_vector.stop_words))
`````

`````python
# Turn our data into vectors
def transform_ua_into_vector(count_vector, user_agents):
    words = count_vector.get_feature_names_out() # the original words behind the encoded data
    data = pd.DataFrame(count_vector.transform(user_agents).toarray())
    user_agents_transformed = data.rename(columns=lambda i: words[i])    
    return user_agents_transformed
`````

`````python
# Fit the training data and then return the matrix
training_data = transform_ua_into_vector(count_vector, X_train)

# Transform testing data and return the matrix. Note we are not fitting the testing data into the CountVectorizer()
testing_data = transform_ua_into_vector(count_vector, X_test)
`````

### Overfit?

`````python
# https://stackoverflow.com/questions/51397109/prune-unnecessary-leaves-in-sklearn-decisiontreeclassifier
from sklearn.tree._tree import TREE_LEAF, TREE_UNDEFINED

def is_leaf(inner_tree, index):
    # Check whether node is leaf node
    return (inner_tree.children_left[index] == TREE_LEAF and 
            inner_tree.children_right[index] == TREE_LEAF)

def prune_index(inner_tree, decisions, index=0):
    # Start pruning from the bottom - if we start from the top, we might miss
    # nodes that become leaves during pruning.
    # Do not use this directly - use prune_duplicate_leaves instead.
    if not is_leaf(inner_tree, inner_tree.children_left[index]):
        prune_index(inner_tree, decisions, inner_tree.children_left[index])
    if not is_leaf(inner_tree, inner_tree.children_right[index]):
        prune_index(inner_tree, decisions, inner_tree.children_right[index])

    # Prune children if both children are leaves now and make the same decision:     
    if (is_leaf(inner_tree, inner_tree.children_left[index]) and
        is_leaf(inner_tree, inner_tree.children_right[index]) and
        (decisions[index] == decisions[inner_tree.children_left[index]]) and 
        (decisions[index] == decisions[inner_tree.children_right[index]])):
        # turn node into a leaf by "unlinking" its children
        inner_tree.children_left[index] = TREE_LEAF
        inner_tree.children_right[index] = TREE_LEAF
        inner_tree.feature[index] = TREE_UNDEFINED
        ##print("Pruned {}".format(index))

def prune_duplicate_leaves(mdl):
    # Remove leaves if both 
    decisions = mdl.tree_.value.argmax(axis=2).flatten().tolist() # Decision for each node
    prune_index(mdl.tree_, decisions)
`````
